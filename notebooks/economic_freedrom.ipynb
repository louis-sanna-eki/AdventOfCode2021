```python
import matplotlib.pyplot as plt


# Fonction pour créer et enregistrer un graphique avec Matplotlib
def create_and_save_plot(x_data, y_data, title, xaxis_title, yaxis_title, filename, autorange=None):
    plt.figure(figsize=(10, 6))
    plt.plot(x_data, y_data, marker='o')
    plt.title(title)
    plt.xlabel(xaxis_title)
    plt.ylabel(yaxis_title)

    if autorange == 'reversed':
        plt.gca().invert_yaxis() # Inverser l'axe Y pour les graphiques où c'est pertinent

    plt.savefig(f"/content/drive/MyDrive/lr-data/{title}.png")
    plt.plot()

# Création et enregistrement de chaque graphique

# Évolution de la note de la France
create_and_save_plot(
    df["Année"], df["Note"],
    "Évolution de la note de la France",
    "Année", "Note",
    "evolution_note_france"
)

# Évolution du rang de la France
create_and_save_plot(
    df["Année"], df["Rang"],
    "Évolution du rang de la France",
    "Année", "Rang",
    "evolution_rang_france",
    autorange='reversed' # Pour le rang, l'autorange est inversé
)

# Note sur la taille de l'État
create_and_save_plot(
    df["Année"], df["Taille de l'État"],
    "Note sur la taille de l'État",
    "Année", "Taille de l'État",
    "note_taille_etat"
)

# Note sur le système légal de protection du droit de propriété
create_and_save_plot(
    df["Année"], df["Droits de propriété"],
    "Note sur le système légal de protection du droit de propriété",
    "Année", "Droits de propriété",
    "note_droits_propriete"
)

# Note sur la qualité de la monnaie
create_and_save_plot(
    df["Année"], df["Qualité de la monnaie"],
    "Note sur la qualité de la monnaie",
    "Année", "Qualité de la monnaie",
    "note_qualite_monnaie"
)

# Note sur la liberté de commercer à l'international
create_and_save_plot(
    df["Année"], df["Liberté de commerce international"],
    "Note sur la liberté de commercer à l'international",
    "Année", "Liberté de commerce international",
    "note_liberte_commerce"
)

# Note sur la réglementation
create_and_save_plot(
    df["Année"], df["Réglementation"],
    "Note sur la réglementation",
    "Année", "Réglementation",
    "note_reglementation"
)
```

```python
import plotly.graph_objs as go

# Fonction pour créer un graphique à barres et enregistrer en tant qu'image PNG
def create_bar_chart(df, x, y, title, ylabel):
    plt.figure(figsize=(10, 8))
    df = df.sort_values(by=y, ascending=True)  # Trier les données pour une meilleure visualisation
    plt.bar(df[x], df[y], color='#1f77b4')

    # Marquer la France en rouge
    if 'France' in df[x].values:
        plt.bar(df[df[x] == 'France'][x], df[df[x] == 'France'][y], color='red')

    plt.title(title, fontsize=16)
    plt.xlabel(x, fontsize=12)
    plt.ylabel("Note", fontsize=12)
    plt.xticks(rotation=90, fontsize=10)
    plt.tight_layout()  # Ajuster automatiquement les paramètres de la figure pour donner de l'espace aux étiquettes

    # Enregistrer l'image
    plt.savefig(f"/content/drive/MyDrive/lr-data/{title}.png")
    plt.plot()

# Créer et enregistrer le graphique de la note par pays
create_bar_chart(df_note, 'Pays', 'Note', 'Note par pays et l\'indice de liberté économique en 2021', 'note_par_pays')

# Créer et enregistrer le graphique de la taille de l'État par pays
create_bar_chart(df_government_size, 'Pays', 'Taille de l\'État', 'Note par pays sur la taille de l\'État en 2021', 'taille_de_etat_par_pays')

# Créer et enregistrer le graphique du système légal de protection du droit de propriété par pays
create_bar_chart(df_legal_system, 'Pays', 'Droits de propriété', 'Note par pays sur le système légal de protection du droit de propriété en 2021', 'droit_de_propriete_par_pays')

create_bar_chart(df_sound_money, 'Pays', 'Sound Money',
                               "Note par pays sur la qualité de la monnaie en 2021",
                              'qualite_monnaie_par_pays')

# Freedom to trade internationally
create_bar_chart(df_trade_freedom, 'Pays', 'Freedom to Trade Internationally',
                               "Note par pays sur la liberté de commercer à l'international en 2021",
                               'liberte_commerce_par_pays')

# Regulation
create_bar_chart(df_regulation, 'Pays', 'Regulation',
                               "Note par pays sur la réglementation en 2021",
                               'reglementation_par_pays')
```
